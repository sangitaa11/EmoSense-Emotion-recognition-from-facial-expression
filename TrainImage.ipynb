```python
import pandas as pd
import numpy as np
import os
import cv2

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/GEN-AI/train.csv', delimiter=',')
```

```python
df.info()
```

```python
# Define emotion labels
emo_dict = {0: "Angry", 1: "Disgust", 2: "Fear", 3: "Happy", 4: "Sad", 5: "Surprise", 6: "Neutral"}

# Create directories for each emotion in the Training directory in your Google Drive folder
drive_folder = '/content/drive/MyDrive/GEN-AI/training'
os.makedirs(drive_folder, exist_ok=True)
for emo_label in emo_dict.values():
    os.makedirs(os.path.join(drive_folder, emo_label), exist_ok=True)
```

```python
# Function to save images
def save_fer_img():
    for index, row in df.iterrows():
        pixels = np.asarray(list(row['pixels'].split(' ')), dtype=np.uint8)
        emotion = row['emotion']
        img = pixels.reshape((48, 48))
        pathname = os.path.join(drive_folder, emo_dict[emotion], f"image_{index}.jpg")
        cv2.imwrite(pathname, img)
        print(f"Image {index} saved at: {pathname}")

# Call the function to save images
save_fer_img()
```
