```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the train data from train.csv
train_df = pd.read_csv('/content/drive/MyDrive/GEN-AI/train.csv')

# Function to convert string of pixel values to numpy array
def string_to_pixels(pixel_string):
    pixels = np.array(pixel_string.split(), dtype=np.float32)
    return pixels

# Extract pixel values and labels for train data
train_pixels = train_df['pixels'].apply(string_to_pixels).values / 255.0  # Normalize pixel values
train_labels = train_df['emotion'].values

# Plot a sample of the train data
plt.figure(figsize=(10, 10))
for i in range(25):  # Plot first 25 images
    plt.subplot(5, 5, i + 1)
    plt.imshow(train_pixels[i].reshape(48, 48), cmap='gray')  # Reshape for plotting
    plt.title(f"Label: {train_labels[i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()

```
