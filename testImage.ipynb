```python
import pandas as pd
import os
import numpy as np
import cv2


# Create directories for each emotion category within the "Testing" directory in Google Drive
os.mkdir("/content/drive/MyDrive/GEN-AI/testing/Angry")
os.mkdir("/content/drive/MyDrive/GEN-AI/testing/Disgust")
os.mkdir("/content/drive/MyDrive/GEN-AI/testing/Fear")
os.mkdir("/content/drive/MyDrive/GEN-AI/testing/Happy")
os.mkdir("/content/drive/MyDrive/GEN-AI/testing/Sad")
os.mkdir("/content/drive/MyDrive/GEN-AI/testing/Surprise")
os.mkdir("/content/drive/MyDrive/GEN-AI/testing/Neutral")

# Define the dictionary mapping emotion labels to their corresponding directories
emo_dict = {0: "Angry", 1: "Disgust", 2: "Fear", 3: "Happy", 4: "Sad", 5: "Surprise", 6: "Neutral"}

# Load testing data from the CSV file
fer_data = pd.read_csv('/content/drive/MyDrive/GEN-AI/test.csv')
```

```python
fer_data.info()
```

```python
# Iterate through the testing data and save images to appropriate directories
def save_fer_img():
    for index, row in fer_data.iterrows():
        # Extract emotion label and pixel values from the current row
        emotion_label = row['emotion']
        pixels = np.asarray(list(row['pixels'].split(' ')), dtype=np.uint8)

        # Reshape pixel values into an image
        img = pixels.reshape((48, 48))

        # Determine the directory path based on the emotion label
        directory_path = os.path.join("/content/drive/MyDrive/GEN-AI/testing", emo_dict[emotion_label])

        # Create the directory if it doesn't exist
        if not os.path.exists(directory_path):
            os.mkdir(directory_path)

        # Save the image in the appropriate directory
        img_filename = f"{emo_dict[emotion_label]}_{index}.jpg"
        img_path = os.path.join(directory_path, img_filename)
        cv2.imwrite(img_path, img)
        print(f"Saved image {index} to {img_path}")

# Call the function to save images
save_fer_img()
```
